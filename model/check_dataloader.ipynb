```python
import numpy as np
from dataloaders import GetSingleStepDataLoaders
```

```python
batch_size = 32
```

```python
data  = np.load('../data/test.npz')
c_t = data['coords_t']
c_tp1 = data['coords_tp1']
steps = data['steps']
positions = data['positions']
rotations = data['rotations']
actions = np.hstack((steps, positions, rotations))
```

```python
train_loader, test_loader = GetSingleStepDataLoaders(
    coords_t=c_t,       
    coords_tp1=c_tp1,
    actions=actions,
    batch_size=batch_size
)
```

```python
for i, batch in enumerate(train_loader):
    state_t, actions, delta_tp1, states_tp1 = batch
    print(f"Batch {i}:")
    print("state_t shape:", state_t.shape)
    print("actions shape:", actions.shape)
    print("delta_tp1 shape:", delta_tp1.shape)
    print("states_tp1 shape:", states_tp1.shape)
    
    if i == 2:   # show first 3 batches only
        break
```
