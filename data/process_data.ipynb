```python
import numpy as np
import pyvista as pv
from process_data import extract_data
```

```python
def visualize_point_cloud(pc, point_size=5):
    cloud = pv.PolyData(pc)
    plotter = pv.Plotter()
    plotter.add_points(cloud, point_size=point_size, color="red")
    plotter.show_axes()
    plotter.show()
```

```python
total_points = 2048
lines = 900
db_path = db_path = './RL_random_cogging.db'
```

```python
c_t, c_tp1, s, p, r, series_lengths, series_ids = extract_data(db_path, total_points, lines)
```

```python
print(c_t.shape), print(c_tp1.shape), print(s.shape), print(p.shape), print(r.shape)
print(series_lengths)
print(series_ids)
```

```python
visualize_point_cloud(c_t[0], point_size=5)
```

```python
visualize_point_cloud(c_tp1[0], point_size=5)
```
